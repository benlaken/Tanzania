## 1. Call modules

```python
%matplotlib inline
# Import some required modules and settings
import numpy as np
import scipy as sci
import statsmodels.api as sm
from scipy.stats import pearsonr
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
import matplotlib as mpl

mpl.rcParams['figure.figsize']=(10.0,8.0)
```

```python
# A custom function to produce a two-pannel correlelogram and save it as a pdf
def Correl_plot(precip,precip_error,forest,forest_error,grass,grass_error,lnum):
    '''
    Produce a correlogram of precipitation vs EVI anomalies
    '''
    fname = 'Lag_'+lnum
    #print 'Producing:',fname
    GT_desz_lin = stats.linregress(precip, grass)      # Calculate lin regression
    FT_desz_lin = stats.linregress(precip, forest)
    
    #slope,intercept,rval,pval,stderr = stats.linregress(Precip_deseaz, GT_desz)
    print 'Precip--GT:  rval=',GT_desz_lin[2],'pval=',GT_desz_lin[3]
    print 'Precip--FT:  rval=',FT_desz_lin[2],'pval=',FT_desz_lin[3]    
    x_rng = [min(precip),max(precip)]
    ylin_FT=[i * FT_desz_lin[0] + FT_desz_lin[1] for i in x_rng]
    ylin_GT=[i * GT_desz_lin[0] + GT_desz_lin[1] for i in x_rng]
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
    axes[0].errorbar(precip,forest,yerr=forest_error,xerr=precip_error,
                         c='g',fmt='x',alpha=0.5)
    axes[0].plot(x_rng,ylin_FT,'k-')
    axes[0].set_title('Forest-type at lag '+lnum)
    axes[0].grid(True)
    axes[0].set_xlabel('Monthly Precipitation anomaly (mm)')
    axes[0].set_ylabel('Monthly EVI anomaly')
    axes[0].set_xlim(-300,300)
    axes[1].errorbar(precip,grass,yerr=grass_error,xerr=precip_error,
                         c='r',fmt='x',alpha=0.5)
    axes[1].plot(x_rng,ylin_GT,'k-')
    axes[1].set_title('Grass-type at lag '+lnum)
    axes[1].grid(True)
    axes[1].set_xlabel('Monthly Precipitation anomaly (mm)')
    axes[1].set_ylabel('Monthly EVI anomaly')
    axes[1].set_xlim(-300,300)
    plt.show(fig)
    fig.savefig(fname+'.pdf',dpi=300)
    return
```

## 2. Read data from online files

```python
#Read in the CSV files, automatically create date objects, and remove the original date column (leaving it only as an index)
MODIS = pd.read_csv('http://files.benlaken.com/documents/MODIS_EVI.csv', parse_dates=['Date'])
MODIS.index = MODIS.Date
MODIS = MODIS.drop('Date', 1) 

# Reseample data to monthly resolution
modis_monthly = MODIS.resample('M',how='mean')
```

```python
# Read the data file as a Pandas object (formatted as CSV text file)
tnz_dat = pd.read_csv('http://files.benlaken.com/documents/Iddi_precip.csv')
tnz_dat.index = pd.date_range('19700101', '20111231', freq='M')      # Set a pd.timeobject as an index
tmp = np.where(tnz_dat.Pemba_Airport < -10)                          # Find all points where missing value
tnz_dat.Pemba_Airport[tmp[0]] = np.nan                               # Overwrite -99.0 with NAN
tnz_dat = tnz_dat.drop('mm', 1)                                      # Remove the unwanted data columns
tnz_dat = tnz_dat.drop('yyyy', 1)
sites=list(tnz_dat)                                                  # List object of the site names
```

```python
# Replace missing (-99.) values with np.NAN (so later analysis will function correctly)
for n in sites:
    tmp = tnz_dat.loc[:,n] == -99
    print len(tnz_dat.loc[tmp,n]),'missing values at',n     # Replace -99. missing data with nan
    if len(tnz_dat.loc[tmp,n]) > 0:
        tnz_dat.loc[tmp,n] = np.nan
    
                                                                     # Good_pos= index for every row, of the stations with data
good_pos = [np.where(tnz_dat.loc[tnz_dat.index[rows],sites[:]] >= 0.0) for rows in xrange(len(tnz_dat))]
qmean =[]
qsem = []
for row in xrange(len(tnz_dat)):
    qcsum = [tnz_dat.loc[tnz_dat.index[row],sites[n]] for n in good_pos[row][0]]
    qmean.append(np.mean(qcsum))
    qsem.append(np.std(qcsum)/np.sqrt(len(qcsum) -1))
```

## Process the precipitation values

```python
# Calculate a seasonal climatology of monthly precipitation and subtract it from the time-series
# Nb. this accumulates means using a truth mask and calculates the accumulates Standard error,
# using the formlaue of SEM_accumulated = SQRT((SEM1)^2 + (SEM2)^2 + ..SEMn)

Precip=pd.Series(qmean,index=tnz_dat.index)
Precip_SEM= pd.Series(qsem,index=tnz_dat.index)

Precip_2000 = Precip['2000-01-31':]
PrecipSEM_2000 = Precip_SEM['2000-01-31':]

sclimo = [np.mean(Precip_2000[Precip_2000.index.month == (mnth+1)]) for mnth in xrange(12)]

sclimo_err = [np.sqrt(np.sum(PrecipSEM_2000[PrecipSEM_2000.index.month == (mnth+1)] * 
                            PrecipSEM_2000[PrecipSEM_2000.index.month == (mnth+1)])) 
                            for mnth in xrange(12)]

#Deseasonalize precipitation data
deseaz = [i - sclimo[(Precip_2000.index[n].month)-1] for n,i in enumerate(Precip_2000)]
Precip_deseaz = pd.Series(deseaz,index=Precip_2000.index)   # This is the key dataset (deaseazonalized precip with time index)
```

```python
lags = 24    # <--------------------------------- Set the window size of the lagged analysis here
show = 1     # <--------------------------------- Optionally produce an extra correlelogram for a specific lag

nmax = len(FT_desz)
#print nmax
istart=[]
# This sets the indexes for the dependent series...
zeros = [0] * lags
dend=list(np.arange(nmax - lags,nmax,1))
dend.append(nmax)
for n in zeros:
    dend.append(n +nmax)
dstart=zeros
for n in np.arange(0,lags+1,1):
    dstart.append(n)

# This sets the indexes for the independent series...
zeros = [0] * lags   
istart=list(np.arange(lags,0,-1))
istart.append(zeros[0])
for n in zeros:
    istart.append(n)
iend=[nmax]*(lags+1)
for n in np.arange(nmax-1,nmax - (lags+1),-1):
    iend.append(n)

prft_rvals=[]
prft_pvals=[]
prft_slope=[]
prft_err=[]
prgt_rvals=[]
prgt_pvals=[]
prgt_slope=[]
prgt_err=[]
lag_list = []

print 'Generating Lagged Regression Statistics...'
for n in xrange(len(dstart)):
    #print 'Lag',n - (lags)
    Pr_ind = Precip_deseaz[istart[n]:iend[n]]
    Pr_err = PrecipSEM_2000[istart[n]:iend[n]] 
    FT_dep = FT_desz[dstart[n]:dend[n]]
    FT_err = FT_desz_err[dstart[n]:dend[n]]
    GT_dep = GT_desz[dstart[n]:dend[n]]
    GT_err = GT_desz_err[dstart[n]:dend[n]]
    #print 'Lag',n - (lags),'D. len:',len(FT_dep),'start-end',dstart[n],dend[n],  
    #print 'I. len:',len(Pr_ind),' start-end',istart[n],iend[n] 
    
    slope,intercept,rval,pval,stderr = stats.linregress(Pr_ind, FT_dep)
    prft_rvals.append(rval)
    prft_pvals.append(pval)
    prft_slope.append(slope)
    prft_err.append(stderr)

    slope,intercept,rval,pval,stderr = stats.linregress(Pr_ind, GT_dep)
    prgt_rvals.append(rval)
    prgt_pvals.append(pval)
    prgt_slope.append(slope)
    prgt_err.append(stderr)
    
    if(n-lags == (show)):
        print 'Also producing a plot for Lag:',n-lags
        Correl_plot(Pr_ind,Pr_err,FT_dep,FT_err,GT_dep,GT_err,str(n-lags))
    lag_list.append(n - lags)
```

```python
# Lagged Regression statistics plot
fig2, axes2 = plt.subplots(nrows=4, ncols=1, figsize=(15,10))

axes2[0].errorbar(lag_list, np.array(prft_slope)*10000,xerr=None,yerr=np.array(prft_err)*10000,
                  c='g',alpha=0.8)   # These are the slope values
axes2[0].errorbar(lag_list, np.array(prgt_slope)*10000,xerr=None,yerr=np.array(prgt_err)*10000,
                  c='r',alpha=0.8)
axes2[0].set_ylabel(r'Slope (EVI$\times 10^{-4}$)')
axes2[0].set_ylim(-2,3)
axes2[0].set_xlim(min(lag_list),max(lag_list))
axes2[0].grid(True)

axes2[1].plot(lag_list, prft_rvals,'gx-')   # These are the r-values
axes2[1].plot(lag_list, prgt_rvals ,'rx-')
axes2[1].set_ylabel('$r$-values')
axes2[1].set_xlim(min(lag_list),max(lag_list))
axes2[1].grid(True)

axes2[2].plot(lag_list, (np.array(prft_rvals) *np.array(prft_rvals)),'gx-')   # These are the r^2-values
axes2[2].plot(lag_list, (np.array(prgt_rvals) * np.array(prgt_rvals)) ,'rx-') # (Coefficent of determination)
axes2[2].set_ylabel('$r^{2}$-values')
axes2[2].set_xlim(min(lag_list),max(lag_list))
axes2[2].grid(True)

axes2[3].plot(lag_list, prft_pvals,'gx-')  # These are the p-values
axes2[3].plot(lag_list, prgt_pvals,'rx-')
axes2[3].set_ylabel('$p$-values')
axes2[3].grid(True)
axes2[3].set_xlim(min(lag_list),max(lag_list))
axes2[3].set_xlabel('Lag (months)')

plt.show(fig2)    
fig2.savefig('Stat_Analysis.pdf',dpi=300)
```

```python
lags = 24    # <--------------------------------- Set the window size of the lagged analysis here
show = 1     # <--------------------------------- Optionally produce an extra correlelogram for a specific lag

nmax = len(FT_desz)
#print nmax
istart=[]
# This sets the indexes for the dependent series...
zeros = [0] * lags
dend=list(np.arange(nmax - lags,nmax,1))
dend.append(nmax)
for n in zeros:
    dend.append(n +nmax)
dstart=zeros
for n in np.arange(0,lags+1,1):
    dstart.append(n)

# This sets the indexes for the independent series...
zeros = [0] * lags   
istart=list(np.arange(lags,0,-1))
istart.append(zeros[0])
for n in zeros:
    istart.append(n)
iend=[nmax]*(lags+1)
for n in np.arange(nmax-1,nmax - (lags+1),-1):
    iend.append(n)

sz_prft_rvals=[]
sz_prft_pvals=[]
sz_prft_slope=[]
sz_prft_err=[]
sz_prgt_rvals=[]
sz_prgt_pvals=[]
sz_prgt_slope=[]
sz_prgt_err=[]
sz_lag_list = []

print 'Generating Lagged Regression Statistics...'
for n in xrange(len(dstart)):
    #print 'Lag',n - (lags)
    Pr_ind = Pr_szintense[istart[n]:iend[n]]
    Pr_err = PrecipSEM_2000[istart[n]:iend[n]] 
    FT_dep = FT_szintense[dstart[n]:dend[n]]
    FT_err = FT_desz_err[dstart[n]:dend[n]]
    GT_dep = GT_szintense[dstart[n]:dend[n]]
    GT_err = GT_desz_err[dstart[n]:dend[n]]
    #print 'Lag',n - (lags),'D. len:',len(FT_dep),'start-end',dstart[n],dend[n],  
    #print 'I. len:',len(Pr_ind),' start-end',istart[n],iend[n] 
    
    slope,intercept,rval,pval,stderr = stats.linregress(Pr_ind, FT_dep)
    sz_prft_rvals.append(rval)
    sz_prft_pvals.append(pval)
    sz_prft_slope.append(slope)
    sz_prft_err.append(stderr)

    slope,intercept,rval,pval,stderr = stats.linregress(Pr_ind, GT_dep)
    sz_prgt_rvals.append(rval)
    sz_prgt_pvals.append(pval)
    sz_prgt_slope.append(slope)
    sz_prgt_err.append(stderr)
    
    #if(n-lags == (show)):
    #    print 'Also producing a plot for Lag:',n-lags
    #    Correl_plot(Pr_ind,Pr_err,FT_dep,FT_err,GT_dep,GT_err,str(n-lags))
    sz_lag_list.append(n - lags)
```

```python
# Lagged Regression statistics plot
fig3, axes2 = plt.subplots(nrows=4, ncols=1, figsize=(15,10))

axes2[0].errorbar(sz_lag_list, np.array(sz_prft_slope)*10000,xerr=None,yerr=np.array(sz_prft_err)*10000,
                  c='g',alpha=0.8)   # These are the slope values
axes2[0].errorbar(sz_lag_list, np.array(sz_prgt_slope)*10000,xerr=None,yerr=np.array(sz_prgt_err)*10000,
                  c='r',alpha=0.8)
axes2[0].set_ylabel(r'Slope (EVI$\times 10^{-4}$)')
#axes2[0].set_ylim(-2,3)
axes2[0].set_xlim(min(sz_lag_list),max(sz_lag_list))
axes2[0].grid(True)

axes2[1].plot(sz_lag_list, sz_prft_rvals,'gx-')   # These are the r-values
axes2[1].plot(sz_lag_list, sz_prgt_rvals ,'rx-')
axes2[1].set_ylabel('$r$-values')
axes2[1].set_xlim(min(sz_lag_list),max(sz_lag_list))
axes2[1].grid(True)

axes2[2].plot(sz_lag_list, (np.array(sz_prft_rvals) *np.array(sz_prft_rvals)),'gx-')   # These are the r^2-values
axes2[2].plot(sz_lag_list, (np.array(sz_prgt_rvals) * np.array(sz_prgt_rvals)) ,'rx-') # (Coefficent of determination)
axes2[2].set_ylabel('$r^{2}$-values')
axes2[2].set_xlim(min(sz_lag_list),max(sz_lag_list))
axes2[2].grid(True)

axes2[3].plot(sz_lag_list, sz_prft_pvals,'gx-')  # These are the p-values
axes2[3].plot(sz_lag_list, sz_prgt_pvals,'rx-')
axes2[3].set_ylabel('$p$-values')
axes2[3].grid(True)
axes2[3].set_xlim(min(sz_lag_list),max(sz_lag_list))
axes2[3].set_xlabel('Lag (months)')

plt.show(fig3)    
fig3.savefig('Seasonal_Stat_Analysis.pdf',dpi=300)
```

## Conclusions

Data is separated into a regular (seasonal) component, using a DOY mean approach, and the
deseasonalized anomaly.

For the seasonal variations, precipitation observations from coastal sites of Tanzania show
a bi-modal distribution. One strong peak centered on April, another weaker peak on December.

Enhanced Vegetation Index (EVI) measurements from a comparable region (Eastern Tanzania), 
distinguished by areas classified as Grass-land-type, and Forest-land-type (based on a MODIS
land classification from 2009), show a single peak (which begins with the increase associated
with the December weak-peak, and reaches a maximum with the strong April precipitation peak).
Grass-type values change from ~0.50 to 0.21 (indicating a die-off to bare ground), Forest-type
values change from ~0.53 to 0.34 (indicating some perennial retention of vegetation).

Correlation analysis of the seasonal component shows that for every 1 mm change in 
precipitation, positively correlated EVI changes of 6x10-4 and 4x10-4 occur over Forest-
and Grass-type land cover respectively (zero-lag), with a significance of p<0.01.
These changes possess r2 values of 0.46 and 0.41 for Forest- and Grass-type respectively.

The strength of the correlation is somewhat asymmetrical with lag, approximately oscillating every
6-months, this is a result of the bi-modal precipitation peak (which is unevenly spaced in time).

Analysis of the deseasonalized anomaly, which essentially shows the monthly deviations from regular
seasonal values, shows the strongest precipitation---EVI relationship at lag+1: for every 1mm change
in precipitation, there is a positively correlated EVI changes of 2.5x10-4 for Grass-type and 2x10-4 
for Forest-type. R2 values at lag +1 are 0.17 and 0.13 respectively, and also possess a high statistical
significance (p<0.01). It is worth noting for context that the lag +1 seasonal correlations are
higher in terms of absolute r2 value than correlations to the anomaly at lag +1.

Precipitation strongly determines EVI: Forest- and Grass-type show a total r2 of 0.60 and 0.58 respectively, 
i.e. ~60% of monthly EVI changes are positively correlated to Precipitation changes.
The forest EVI appears to be more impacted by the seasonal rains, while the grass-lands are 
relatively more subject to deviations from seasonal norms.
