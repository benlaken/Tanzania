## Read in data and place np.nan in missing data

```python
# Read the data file as a Pandas object (formatted as CSV text file)
tnz_dat = pd.read_csv('http://files.benlaken.com/documents/Iddi_precip.csv')
```

```python
# Set a Pandas time index for the object
tnz_dat.index = pd.date_range('19700101', '20111231', freq='M')   # Set a pd.timeobject as an index
```

```python
tmp = np.where(tnz_dat.Pemba_Airport < -10)                          # Find all points where missing value
tnz_dat.Pemba_Airport[tmp[0]] = np.nan                               # Overwrite -99.0 with NAN

tnz_dat = tnz_dat.drop('mm', 1)  # Remove the unwanted data columns
tnz_dat = tnz_dat.drop('yyyy', 1)
```

```python
sites=list(tnz_dat)  # List object of the site names
print sites
```

```python
# Replace -99. missing data with nan
for n in sites:
    tmp = where(tnz_dat.loc[:,n] < -1)
    print len(tmp[0]),'missing values at',n
    tnz_dat.loc[tmp[0],n] = np.nan 
```

## Calculate Accumulated mean and SEM values

```python
# Good_pos is an index for every row, of the stations with data
good_pos = [np.where(tnz_dat.loc[tnz_dat.index[rows],sites[:]] >= 0.0) for rows in xrange(len(tnz_dat))]
qmean =[]
qsem = []
for row in xrange(len(tnz_dat)):
    qcsum = [tnz_dat.loc[tnz_dat.index[row],sites[n]] for n in good_pos[row][0]]
    qmean.append(np.mean(qcsum))
    qsem.append(np.std(qcsum)/np.sqrt(len(qcsum) -1))

#qmean = pd.Series(qmean,index=tnz_dat.index, name = 'Accumulated')   # Make the mean and SEM lists into pd objects
#qsem = pd.Series(qsem,index=tnz_dat.index, name = 'SEM')

tnz_dat['Accumulated']=pd.Series(qmean,index=tnz_dat.index)  # Mad how easy it is to add columns to the dataframe!
tnz_dat['SEM']= pd.Series(qsem,index=tnz_dat.index)
```

## Seasonal Climatoloty of Precipitation

```python
#Very easy to operate on truth masks of the Dataframe object...
#aaa = tnz_dat.index.month == 1
#print np.mean(tnz_dat.Accumulated[aaa])


sclimo = [np.mean(tnz_dat.Accumulated[tnz_dat.index.month == (mnth+1)]) for mnth in xrange(12)]

sclimo_err = [np.sqrt(np.sum(tnz_dat.SEM[tnz_dat.index.month == (mnth+1)] * 
                            tnz_dat.SEM[tnz_dat.index.month == (mnth+1)])) 
                            for mnth in xrange(12)]

# Nb. this accumulates means using a truth mask
# And also calculates the accumulates Standard error, using the formlaue of
# SEM_accumulated = SQRT((SEM1)^2 + (SEM2)^2 + ..SEMn)
```

## Annual Means

```python
yrange = arange(min(tnz_dat.index.year),max(tnz_dat.index.year))

ann = [np.sum(tnz_dat.Accumulated[tnz_dat.index.year == yr]) for yr 
       in yrange]

ann_err = [np.sqrt(np.sum(tnz_dat.SEM[tnz_dat.index.year == yr] * 
                            tnz_dat.SEM[tnz_dat.index.year == yr])) 
                            for yr in yrange]
```

```python
def runningMeanFast(x, N):
    return np.convolve(x, np.ones((N,))/N)[(N-1):]

len = 5
ann_rmean = runningMeanFast(ann,5)
ann_rmean[-5:] = np.nan
```

```python
my_annual = plt.figure()
my_annual.set_size_inches(10,5)        # Specify the output size
ax1 = my_annual.add_subplot(111)

ax1.errorbar(yrange,ann,xerr=None,yerr=ann_err,c='b',alpha=0.8)
ax1.plot(yrange,ann_rmean,'r-')
plt.ylabel(r'Precipitation (mm yr$^{-1}$)')
plt.xlabel('Year')
#plt.xlim(0,11)
#plt.xticks(arange(12),mnths)
plt.title(r'Annual precipitation recorded in coastal Tanzania (& 5yr smooth)')
plt.grid(True)
plt.show(my_sclimo)

my_annual.savefig('Annual_ts.pdf',dpi=300)
```
